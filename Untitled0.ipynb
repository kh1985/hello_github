```
!pip install --upgrade -q gspread
#!pip install -U tensorflow-gpu==2.0.0 grpcio

import requests
from bs4 import BeautifulSoup
import re
import json
from google.colab import auth
import gspread
from oauth2client.client import GoogleCredentials
import types
import time
from google.colab import drive
import csv

#スプレッドシート認証処理
auth.authenticate_user()
gc = gspread.authorize(GoogleCredentials.get_application_default())

#ティッカー一覧を取得
ss_url = "初心者のための投資講座"
workbook = gc.open(ss_url)
worksheet = workbook.worksheet("ticker_list")
#cell_list = worksheet.range("A3:A3865")
cell_list = []
for cell in worksheet.range("A3:A90"):
  cell_list.append(cell.value)

class Settlement:
  BASE_FINANCIALS_URL='https://finance.yahoo.com/quote/{0}/financials?p={0}'
  BASE_CASHFLOW_URL='https://finance.yahoo.com/quote/{0}/cash-flow?p={0}'
  outPutDict = {}
  errCnt = 0
  waitCounter = 0
  WAIT_COUNT = 40

  def checkKey(target,key,selt):
    if key in target.keys():
      return true
    else:
      return false

  def getJScript(self,url):
    r = requests.get(url)
    
    html = r.text
    soup = BeautifulSoup(html,'html.parser')
    soup_script = soup.find("script",text=re.compile("root.App.main")).text
    return json.loads(re.search("root.App.main\s+=\s+(\{.*\})",soup_script)[1])
  
  def getData(self,i):
    try:
      fUrl = self.BASE_FINANCIALS_URL.format(cell_list[i])
      fJScript = self.getJScript(fUrl)
      fQSS = fJScript['context']['dispatcher']['stores']['QuoteSummaryStore']

      if fQSS['cashflowStatementHistory'] is None or len(fQSS['cashflowStatementHistory']) == 0:
        print("cashflowStatementHistory:None")
        return None
      else:
        fCFS = fQSS['cashflowStatementHistory']['cashflowStatements']
      fISH = fQSS['incomeStatementHistory']['incomeStatementHistory']

      #if fJScript['context']['dispatcher']['stores']['QuoteTimeSeriesStore']['timeSeries'] is None:
      #  print("timeSeries:None")
      #  return None
      #else:
      #  fADAS = fJScript['context']['dispatcher']['stores']['QuoteTimeSeriesStore']['timeSeries']['annualDilutedAverageShares']
      fADAS = fJScript['context']['dispatcher']['stores']['QuoteTimeSeriesStore']['timeSeries']['annualDilutedAverageShares']

      cUrl = self.BASE_CASHFLOW_URL.format(cell_list[i])
      cJScript = self.getJScript(cUrl)
      cQSS = cJScript['context']['dispatcher']['stores']['QuoteSummaryStore']

      cCFS = cQSS['cashflowStatementHistory']['cashflowStatements']
      
      #max3年のデータがほしいけど、0~2も存在する。
      roopCnt = len(fADAS) -1
      if roopCnt < 0:
        return None
      fadasiBaseCnt = roopCnt - 1

      #if cell_list[i] == 'ACET'
      #print("fCFS"+str(len(fCFS)))
      #print("fISH"+str(len(fISH)))
      #print("fADAS"+str(len(fADAS)))
      #print("cCFS"+str(len(cCFS)))
      
      #print("roopCnt:" + str(roopCnt))
      print("end:" + fUrl)

      for j in range(roopCnt):
         fadasi = fadasiBaseCnt-j
         dataDict = {}
         dataDict['endDate'] = fCFS[j]['endDate']['fmt'] if 'endDate' in fCFS[j].keys() else ''
         dataDict['netIncome'] = fCFS[j]['netIncome']['longFmt'] if 'netIncome' in fCFS[j].keys() else ''
         dataDict['totalRevenue'] = fISH[j]['totalRevenue']['longFmt'] if 'totalRevenue' in fISH[j].keys() else ''
         dataDict['reportedValue'] = fADAS[fadasi]['reportedValue']['raw'] if 'reportedValue' in fADAS[fadasi].keys() else ''
         dataDict['totalCashFromOperatingActivities'] = cCFS[j]['totalCashFromOperatingActivities']['longFmt'] if 'totalCashFromOperatingActivities' in fCFS[j].keys() else ''
         dataDict['dividendsPaid'] = cCFS[j]['dividendsPaid']['longFmt'] if 'dividendsPaid' in cCFS[j].keys() else ''

         # カンマ除去
         dataDict['netIncome'] = dataDict['netIncome'].replace(',', '')
         dataDict['totalRevenue'] = dataDict['totalRevenue'].replace(',', '')
         dataDict['totalCashFromOperatingActivities'] = dataDict['totalCashFromOperatingActivities'].replace(',', '')
         dataDict['dividendsPaid'] = dataDict['dividendsPaid'].replace('-', '') #配当は-は除去する
         dataDict['dividendsPaid'] = dataDict['dividendsPaid'].replace(',', '')

         key = cell_list[i] + '_' + dataDict['endDate']
         self.outPutDict[key] = dataDict
    except:
        if self.errCnt < 5:
          print("こけた！")
          self.errCnt = self.errCnt + 1
          self.getData(i)

  def getAllData(self):
    for i in range(len(cell_list)):
      self.errCnt = 0
      self.getData(i)
      self.waitCounter = self.waitCounter +1
      if self.waitCounter % self.WAIT_COUNT == 0:
        print("休憩")
        time.sleep(30)

S = Settlement()
allData = {}
S.getAllData()
allData = S.outPutDict

#drive.mount('/content/drive')
with open('/content/drive/My Drive/settlement.csv', 'a') as f:
  w = csv.writer(f)
  ky = 1
  worksheet_output_list = workbook.worksheet("output_list")
  for symbl,d in allData.items():
    kx = 1
    row = []
    for k,v in d.items():
      if kx == 1:
        row.append(symbl)
      else:
        row.append(v)
      kx += 1
    w.writerow(row)
    ky += 1
```
